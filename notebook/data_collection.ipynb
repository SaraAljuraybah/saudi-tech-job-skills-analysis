```python
import requests
import pandas as pd
import time

url = "https://jsearch.p.rapidapi.com/search"

headers = {
    "x-rapidapi-host": "jsearch.p.rapidapi.com",
    "x-rapidapi-key": "f112a23bd9msh42e08e2a09d9e0ep10fb00jsnc343bce146de"
}

queries = [
    "data scientist",
    "data analyst",
    "machine learning engineer",
    "software engineer",
    "cybersecurity"
]

all_jobs = []

for q in queries:
    params = {
        "query": q,
        "country": "sa",
        "page": 1,
        "num_pages": 3
    }

    response = requests.get(url, headers=headers, params=params)
    data = response.json()
    jobs = data.get("data", [])

    for job in jobs:
        job["search_query"] = q
        all_jobs.append(job)

    time.sleep(1)


df_raw = pd.json_normalize(all_jobs)


df_raw.to_csv("jobs_sa_raw.csv", index=False)

print("Number Of Jobs :", len(df_raw))
df_raw.head()
```
